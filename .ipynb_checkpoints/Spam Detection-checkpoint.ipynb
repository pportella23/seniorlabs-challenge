```python
from sklearn.impute          import SimpleImputer
from sklearn.preprocessing   import StandardScaler
from sklearn.preprocessing   import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.pipeline        import Pipeline
from sklearn.compose         import ColumnTransformer

cols_to_drop = ['Full_Text', 'Date', 'IsSpam', 'Month'] 
colunas_numericas = df.drop(cols_to_drop, axis=1)

atributos_numericos = colunas_numericas.select_dtypes(exclude="object").columns

coluna_alvo = ['IsSpam']

df_treino, df_teste = train_test_split(df, test_size=0.2, random_state=42)

# separando a coluna alvo do conjunto de treino
df_treino_labels = df_treino[coluna_alvo].copy()
df_treino        = df_treino.drop(columns=coluna_alvo)

# separando a coluna alvo do conjunto de teste
df_teste_labels = df_teste[coluna_alvo].copy()
df_teste        = df_teste.drop(columns=coluna_alvo)

# Pipelines de PrÃ©-processamento
pipeline_atr_numericos = Pipeline([
    ('scaler', StandardScaler()),
])

preproc_completo = ColumnTransformer([
    ('numericos',   pipeline_atr_numericos, atributos_numericos   ),
    ], sparse_threshold=0)

# pre-processamento do conjunto de treino
X_treino = preproc_completo.fit_transform(df_treino)

# pre-processamento do conjunto de teste
X_teste = preproc_completo.transform(df_teste)

# pre-processamento da coluna alvo para treno
y_treino = df_treino_labels.values == 'yes'
# pre-processamento da coluna alvo para teste
y_teste = df_teste_labels.values == 'yes'

nomes_atributos = atributos_numericos

print("NÃºmero total de atributos: ", np.size(nomes_atributos))
```

### Grid Search

```python
def MLPKerasTF(hidden_layer_sizes=(20),activation="tanh"):
    model = keras.models.Sequential()
    model.add(keras.layers.InputLayer(input_shape=X_treino.shape[1:]))
    for units in hidden_layer_sizes:
        model.add(keras.layers.Dense(units,activation=activation))
    model.add(keras.layers.Dense(1, activation="sigmoid"))
    model.compile(loss="binary_crossentropy", optimizer="sgd", metrics=["accuracy"])
    model.summary()
    print("activation=",activation)
    return model

mlpKerasTF_clf = keras.wrappers.scikit_learn.KerasClassifier(MLPKerasTF)
```

```python
from sklearn.model_selection import GridSearchCV
import time

param_grid = [
 {'hidden_layer_sizes': [(5,), (10,), (100,), (5,5), (5,10), (10,10), (5,100), (10,10,10), (100, 80)],
  'activation': ['sigmoid', 'tanh', 'relu']},
 ]

grid_search = GridSearchCV(mlpKerasTF_clf, param_grid)

start_time = time.time()
grid_search.fit(X_treino, y_treino, epochs=15)
```

```python
print("Tempo de execuÃ§Ã£o:  %s segundos" % (time.time() - start_time))
```

```python
grid_search.best_params_
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

y_previsto = grid_search.best_estimator_.predict(X_teste)
print("Best MLPKerasTF:")
print("AcurÃ¡cia: ", accuracy_score(y_teste,y_previsto))
print("PrecisÃ£o: ", precision_score(y_teste,y_previsto))
print("Recall:   ", recall_score(y_teste,y_previsto))
print("F1 Score: ", f1_score(y_teste,y_previsto))
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import matplotlib.pyplot as plt

model = keras.models.Sequential()
model.add(keras.layers.Dense(100, activation="tanh",input_shape=X_treino.shape[1:]))
model.add(keras.layers.Dense(80, activation="tanh"))
model.add(keras.layers.Dense(1, activation="sigmoid"))

model.compile(loss="binary_crossentropy", optimizer="sgd", metrics=["accuracy"])

early_stopping = keras.callbacks.EarlyStopping(patience=3, restore_best_weights=True)

history = model.fit(X_treino, y_treino, epochs=10, validation_split=0.2, callbacks=[early_stopping])


y_previsto_rna = model.predict(X_teste)
y_previsto_rna[y_previsto_rna > 0.50]=True
y_previsto_rna[y_previsto_rna<=0.50]=False

print("RNA:")
print("AcurÃ¡cia: ", accuracy_score(y_teste,y_previsto_rna))
print("PrecisÃ£o: ", precision_score(y_teste,y_previsto_rna))
print("Recall:   ", recall_score(y_teste,y_previsto_rna))
print("F1 Score: ", f1_score(y_teste,y_previsto_rna))

pd.DataFrame(history.history).plot(figsize=(10,6))
plt.grid(True)
plt.gca().set_ylim(0,1)
plt.show()
```

### ReduÃ§Ã£o Dimensionalidade

```python
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression

pca = PCA(n_components=10)

X_train_proj = pca.fit_transform(X_treino)
X_test_proj = pca.transform(X_teste)
```

```python
model = keras.models.Sequential()
model.add(keras.layers.Dense(100, activation="tanh",input_shape=X_train_proj.shape[1:]))
model.add(keras.layers.Dense(80, activation="tanh"))
model.add(keras.layers.Dense(1, activation="sigmoid"))

model.compile(loss="binary_crossentropy", optimizer="sgd", metrics=["accuracy"])

early_stopping = keras.callbacks.EarlyStopping(patience=3, restore_best_weights=True)

history = model.fit(X_train_proj, y_treino, epochs=10, validation_split=0.2, callbacks=[early_stopping])


y_pred = model.predict(X_test_proj)

y_pred[y_pred > 0.50]=True
y_pred[y_pred<=0.50]=False

print("RNA:")
print("AcurÃ¡cia: ", accuracy_score(y_teste,y_pred))
print("PrecisÃ£o: ", precision_score(y_teste,y_pred))
print("Recall:   ", recall_score(y_teste,y_pred))
print("F1 Score: ", f1_score(y_teste,y_pred))

pd.DataFrame(history.history).plot(figsize=(10,6))
plt.grid(True)
plt.gca().set_ylim(0,1)
plt.show()
```
